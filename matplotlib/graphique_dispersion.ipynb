```python
import matplotlib.pyplot as plt

# Données
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11] 

x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]

x3 = [1, 2, 3, 4, 5]
y3 = [3, 5,7, 9, 12] 
# Création du graphique
plt.scatter(x1, y1, color="r", label="Données 1")
plt.scatter(x2, y2, color="b", label="Données 2")
plt.scatter(x3, y3, color="g", label="Données 3")

# Ajout de titres et étiquettes
plt.title("Graphique de Dispersion")
plt.xlabel("Titre pour l'axe X")
plt.ylabel("Titre pour l'axe Y")

# Ajout de la légende
plt.legend()

# Affichage du graphique
plt.show()
```
