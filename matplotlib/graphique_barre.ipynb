```python
import matplotlib.pyplot as plt

# Données
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Création du graphique
plt.bar(categories, values,width=0.5)

# Ajout de titres et étiquettes
plt.title("Graphique en Barres")
plt.xlabel("Catégories")
plt.ylabel("Valeurs")

# Affichage du graphique
plt.show()
```
