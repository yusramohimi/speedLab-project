```python
import matplotlib.pyplot as plt

# Données
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]  
y2 = [1, 4, 6, 8, 10]  

# Création du graphique
plt.fill_between(x, y1, color="skyblue", alpha=0.4, label="Zone sous la courbe 1")
plt.fill_between(x, y2, color="salmon", alpha=0.4, label="Zone sous la courbe 2")

# Tracer les courbes
plt.plot(x, y1, color="blue", linestyle="-", linewidth=2, marker="o", label="Courbe 1")
plt.plot(x, y2, color="red", linestyle="-", linewidth=2, marker="x", label="Courbe 2")

# Ajout de titres et étiquettes
plt.title("Graphique en Courbes de Remplissage")
plt.xlabel("Titre pour l'axe X")
plt.ylabel("Titre pour l'axe Y")

# Ajout de la légende
plt.legend()

# Affichage du graphique
plt.show()
```
